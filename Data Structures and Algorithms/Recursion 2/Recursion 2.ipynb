```python
#Remove X

def removeX(s):
    chr = 'x'
    l = len(s)
    if len(s) == 0:
        return s
    smalleroutput = removeX(s[1:])
    if s[0] == chr :
        return ''+smalleroutput
    else:
        return s[0]+smalleroutput
    
string = input()
print(removeX(string))
```

```python
#Remove Duplicates Recursively

def removeConsecutiveDuplicates(s):
    if len(s)<=1:
        return s
    smallOutput=removeConsecutiveDuplicates(s[1:])
    if s[0]==s[1]:
        return smallOutput
    else:
        return s[0]+smallOutput

# Main
string = input().strip()
print(removeConsecutiveDuplicates(string))
```

```python
#Merge Sort Code

def merge(a1, a2, a):
    i = 0
    j = 0
    k = 0

    while i < len(a1) and j < len(a2):
        if a1[i] < a2[j]:
            a[k] = a1[i]
            k = k + 1
            i = i + 1
        else:
            a[k] = a2[j]
            k = k + 1
            j = j + 1

    while i < len(a1):
        a[k] = a1[i]
        k = k + 1
        i = i + 1

    while j < len(a2):
        a[k] = a2[j]
        k = k + 1
        j = j + 1


def MergeSort(a):
    if len(a) == 0 or len(a) == 1:
        return

    mid = len(a) // 2

    a1 = a[:mid]
    a2 = a[mid:]

    MergeSort(a1)
    MergeSort(a2)

    merge(a1, a2, a)
# a = [5,2,45,12,3,6,52,41,25]


n=int(input())
a=list(int(i) for i in input().strip().split(' '))
MergeSort(a)
print(*a)
```

```python
#Quick sort code

def partition(arr, start, end):
    pivot = arr[start]
    
    count = 0
    for i in range(start, end+1):
        if arr[i]<pivot:
            count += 1
    arr[start+count], arr[start] = arr[start], arr[start+count]
    pivot_index = start+count
    
    i = start
    j = end
    
    while i<j:
        if arr[i]<pivot:
            i += 1
        elif arr[j]>=pivot:
            j -= 1
        else:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
            j -= 1
    return pivot_index

def quickSort(arr, start, end):
    if start>=end:
        return 
    
    pivot_index = partition(arr, start, end)
    quickSort(arr, start, pivot_index-1)
    quickSort(arr, pivot_index+1, end)
      
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
quickSort(arr, 0, n-1)
print(*arr)
```

```python
#Tower of Hanoi


def towerofhanoi(n, source, aux, dest):
    if n==0 :
        return
    
    if n==1 :
        print(source, ' ', dest)
        return
    
    towerofhanoi(n-1, source, dest, aux)
    print(source, ' ', dest)
    towerofhanoi(n-1, aux, source, dest)

n=int(input())
towerofhanoi(n, 'a', 'b', 'c')
```

```python
# Replace all a by b

def replaceChar(string, a ,b):

  if len(string)==0 :
    return string

  if string[0]==a :
    return b + replaceChar(string[1:], a, b)
  else:
    return string[0] + replaceChar(string[1:], a, b)

string = input()
a = input()
b = input()
print(replaceChar(string, a, b))
```

```python
# Replace 'pi' with 3.14

def replacePi(string):
  if len(string)==0 or len(string)==1 :
    return string

  if string[0]=='p' and string[1]=='i' :
    return '3.14' + replacePi(string[2:])
  else:
    return string[0] + replacePi(string[1:])

string = input()
print(replacePi(string))
```

```python
# Binary Search using Recursion

def binarySearch(arr, x, start, end):
  if start>end:
    return -1

  mid = (start+end)//2

  if arr[mid]==x :
    return mid
  elif arr[mid]>x :
    return binarySearch(arr, x, start, mid-1)
  else:
    return binarySearch(arr, x, mid+1, end)

arr = list(int(i) for i in input().strip().split(' '))
x = int(input())
print(binarySearch(arr, x, 0, len(arr)-1))
```
